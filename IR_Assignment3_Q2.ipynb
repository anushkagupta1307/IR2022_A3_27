```python
# importing the necessary libraries..
import networkx as nx
from networkx.drawing.layout import spring_layout
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating the graph..
path = '/content/drive/MyDrive/IR Assignment 3/p2p-Gnutella04.txt'
graph = nx.read_edgelist(path,comments='#',delimiter='\t',create_using=nx.DiGraph)
```

```python
# Getting the number of nodes and edges in the graph..
print(len(graph.nodes()),len(graph.edges()))
```

#PageRank score for each node

```python
# getting the page rank scores for each node..
pgrank_scores = nx.pagerank(graph)
print(len(pgrank_scores))
# ranking the nodes based on their page rank scores..
dict(sorted(pgrank_scores.items(), key=lambda item: item[1], reverse = True))
```

```python
# plotting the Page Rank score distribution..
pgRank_df = pd.DataFrame({"scores":pgrank_scores.values()})
pgRank_df.plot.hist(['scores'], figsize=(8, 6),rwidth=0.8)
plt.xlabel("Page Rank scores")
plt.title("Page Rank Score Distribution")
plt.grid(True)
plt.legend()
```

## Authority and Hub Scores

```python
# getting the Authority scores and Hub scores for each node 
# HITS algorithm is used here..
hub_scores, authority_scores = nx.hits(graph, max_iter = 100, normalized = True)
```

```python
# ranking the nodes based on their Authority scores..
print(len(authority_scores))
dict(sorted(authority_scores.items(), key=lambda item: item[1], reverse = True))
```

```python
# plotting the Authority score distribution
Auth_df = pd.DataFrame({"scores":authority_scores.values()})
Auth_df.plot.hist(['scores'], figsize=(8, 6),rwidth=0.8)
plt.xlabel("Authority scores")
plt.title("Authority Score Distribution")
plt.grid(True)
plt.legend()
```

```python
# ranking the nodes based on their Hub scores..
print(len(hub_scores))
dict(sorted(hub_scores.items(), key=lambda item: item[1], reverse = True))
```

```python
# plotting the Hub score distribution
Hub_df = pd.DataFrame({"scores":hub_scores.values()})
Hub_df.plot.hist(['scores'], figsize=(8, 6),rwidth=0.8)
plt.xlabel("Hub scores")
plt.title("Hub Score Distribution")
plt.grid(True)
plt.legend()
```

```python
# plotting the comparision plot to show comparision of scores of all the three scoring methods..

# getting the nodes ranked according to their scores..
pgRank_sorted = dict(sorted(pgrank_scores.items(), key=lambda item: item[1], reverse = True))
auth_sorted = dict(sorted(authority_scores.items(), key=lambda item: item[1], reverse = True))
hub_sorted = dict(sorted(hub_scores.items(), key=lambda item: item[1], reverse = True))

# getting the highest ranked node for each method..
pgRank_maxKey = int(list(pgRank_sorted.keys())[0])   # node having the highest page rank...
auth_maxKey = int(list(auth_sorted.keys())[0])       # node having the highest authority score...
hub_maxKey = int(list(hub_sorted.keys())[0])         # node having the highest hub score...

# getting the higest score for each method..
pgRank_score = pgRank_sorted[str(pgRank_maxKey)]  # highest page rank score..
auth_score = auth_sorted[str(auth_maxKey)]        # highest authority score..
hub_score = hub_sorted[str(hub_maxKey)]           # highest hub score..

# plotting the figure using the highest ranked node and highest scores obtained above...
plt.figure(figsize=(8, 6))
plt.plot([pgRank_maxKey,auth_maxKey,hub_maxKey],[pgRank_score,auth_score,hub_score],marker='o')
plt.text(pgRank_maxKey,pgRank_score,'   Page Rank Score',color="red")
plt.text(auth_maxKey,auth_score,'   Authority Score',color="red")
plt.text(hub_maxKey,hub_score,'  Hub Score',color="red")
plt.xlabel("Nodes")
plt.ylabel("Scores")
plt.title("Comparision of Scoring Methods")
plt.grid(True)
plt.show()
```
